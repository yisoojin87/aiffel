# 고양이 수염 스티커 만들기

### step2. 얼굴 검출 & 랜드마크 검출하기

```python
# 필요한 라이브러리 import하기

import cv2
import matplotlib.pyplot as plt
import numpy as np
print("완료!!")
```

```python
# opencv로 이미지 읽어오기

import os
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/soojin.png'
img_bgr = cv2.imread(my_image_path)    # OpenCV로 이미지를 읽어온 다음
img_brg = cv2.resize(img_bgr,(640,360))    # 640x360의 크기로 Resize하기
img_show = img_bgr.copy()     # 출력용 이미지 별도 보관
plt.imshow(img_bgr)
plt.show()
```

```python
# RGB 변환

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# hog detector 선언

import dlib
detector_hog = dlib.get_frontal_face_detector()   # detector 선언
print("슝=3")
```

```python
# 얼굴의 bounding box 추출

dlib_rects = detector_hog(img_rgb, 1)
print("슝=3")
```

```python
# 찾은 얼굴을 화면에 출력

print(dlib_rects)  # 찾은 얼굴영역 좌표

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    cv2.rectangle(img_show,(l,t),(r,b),(0,255,0),2,lineType=cv2.LINE_AA)
    
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
# 저장한 landmark모델 불러오기

import os
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
print("슝=3")
```

```python
# landmark 찾기

list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
    
print(len(list_landmarks[0]))
```

```python
# landmark 출력

for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0,255,255), -1)  # yellow
        
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

### step3. 스티커 적용 위치 확인하기

```python
# 좌표 확인하기

for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print(landmark[30])  # 코 중앙의 index는 30
    x = landmark[33][0]
    y = landmark[30][1] 
    w = dlib_rect.width()
    h = dlib_rect.width()
    print('(x,y):(%d,%d)'%(x,y))
    print('(w,h):(%d,%d)'%(w,h))
```

```python
# 준비해 둔 스티커 이미지 읽어오기

import os
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print(img_sticker.shape)
```

```python
# 고양이 수염 스티커 이미지 좌표

refined_left_top_x = x - w//2    # left_top_x
refined_left_top_y = y - h//2    # left_top_y
refined_right_bottom_x = x + w//2   # right_bottom_x
refined_right_bottom_y = y + h//2   # right_bottom_y
print('(left_top_x,left_top_y):(%d,%d)'%(refined_left_top_x, refined_left_top_y))
print('(right_bottom_x,right_bottom_y):(%d,%d)'%(refined_right_bottom_x, refined_right_bottom_y))
```

### step4. 스티커 적용하기

```python
# 원본 이미지에 스티커를 적용

sticker_area = img_show[refined_left_top_y:refined_right_bottom_y, refined_left_top_x:refined_right_bottom_x]
img_show[refined_left_top_y:refined_right_bottom_y, refined_left_top_x:refined_right_bottom_x] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)

# 결과 이미지 출력
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```
